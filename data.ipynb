```python
#라이프데이터만 사용 
life_data = df_without_missing.iloc[:, :4].join(df_without_missing.iloc[:, 373:])
```

```python
# 최근 보험계약고객비율 - 나이 
list = ['la01r', 'la02r', 'la03r', 'la04r', 'la05r','la06r']
dict = {'la01r':'종신', 'la02r':'건강', 'la03r':'상해', 'la04r':'어린이', 'la05r':'연금저축','la06r':'변액'}

for i in list:
    y = i
    name = dict[i]
    print(f"{name} 과 나이 사이에 평균이 유의한지 검정 ")
    age_columns = life_data[['pk3', y]]

    from scipy.stats import f_oneway
    groups = [age_columns[y][age_columns['pk3'] == category] for category in age_columns['pk3'].unique()]

    f_statistic, p_value = f_oneway(*groups)
    print("p-value:", p_value)

    alpha = 0.05  
    if p_value < alpha:
        print("귀무가설을 기각하고 대립가설을 채택합니다. 두 변수 간에는 통계적으로 유의한 차이가 있습니다.")
    else:
        print("귀무가설을 기각하지 못하고 두 변수 간에는 통계적으로 유의한 차이가 없습니다.")

    
    overall_mean = life_data[y].mean()
    group_means = [group.mean() for group in groups]

    for category, mean_diff in zip(life_data['pk3'].unique(), group_means):
        diff_from_overall_mean = mean_diff - overall_mean
        print(f"범주형 변수가 '{category}' 값을 가질 때, 그 그룹은 전체 평균과 {diff_from_overall_mean:.4f} 정도 차이가 납니다.")
```

```python
# 보험금 지급 경험 고객 비율  - 나이 
list = ['lc02r', 'lc03r', 'lc04r', 'lc05r', 'lc06r','lc07r']
dict = {'la01r':'종신', 'la02r':'건강', 'la03r':'상해', 'la04r':'어린이', 'la05r':'연금저축','la06r':'변액'}

for i in list:
    y = i
    name = dict[i]
    print(f"{name} 과 나이 사이에 평균이 유의한지 검정 ")
    age_columns = life_data[['pk3', y]]

    from scipy.stats import f_oneway
    groups = [age_columns[y][age_columns['pk3'] == category] for category in age_columns['pk3'].unique()]

    f_statistic, p_value = f_oneway(*groups)
    print("p-value:", p_value)

    alpha = 0.05  
    if p_value < alpha:
        print("귀무가설을 기각하고 대립가설을 채택합니다. 두 변수 간에는 통계적으로 유의한 차이가 있습니다.")
    else:
        print("귀무가설을 기각하지 못하고 두 변수 간에는 통계적으로 유의한 차이가 없습니다.")

    
    overall_mean = life_data[y].mean()
    group_means = [group.mean() for group in groups]

    for category, mean_diff in zip(life_data['pk3'].unique(), group_means):
        diff_from_overall_mean = mean_diff - overall_mean
        print(f"범주형 변수가 '{category}' 값을 가질 때, 그 그룹은 전체 평균과 {diff_from_overall_mean:.4f} 정도 차이가 납니다.")
```
