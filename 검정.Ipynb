```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
import numpy as np

%matplotlib inline
rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False 
```

```python
data = pd.read_csv('/Users/yoonminseok/Desktop/신빅헤/dat.csv', sep=',')
```

```python
missing_data = data.isna().sum()
print(missing_data)
```

```python
df_without_missing = data.dropna()
```

```python
life_data = df_without_missing.iloc[:, :4].join(df_without_missing.iloc[:, 373:])
```

```python
life_data['성별_연령대'] = life_data['pk2'] + '_' + life_data['pk3']
```

```python
life_data['성별_연령대'].unique()
```

```python
list = ['lc02r', 'lc03r', 'lc04r', 'lc05r', 'lc06r','lc07r']
dict = {'lc02r':'종신', 'lc03r':'건강', 'lc04r':'상해', 'lc05r':'어린이', 'lc06r':'연금저축','lc07r':'변액'}
from scipy.stats import f_oneway

for i in list:
    groups = [life_data[i][life_data['성별_연령대'] == category] for category in life_data['성별_연령대'].unique()]

    f_statistic, p_value = f_oneway(*groups)
    print("p-value:", p_value)

    alpha = 0.05  
    if p_value < alpha:
        print("귀무가설을 기각하고 대립가설을 채택합니다. 두 변수 간에는 통계적으로 유의한 차이가 있습니다.")
    else:
        print("귀무가설을 기각하지 못하고 두 변수 간에는 통계적으로 유의한 차이가 없습니다.")

    overall_mean = life_data[i].mean()
    group_means = [group.mean() for group in groups]

    for category, mean_diff in zip(life_data['성별_연령대'].unique(), group_means):
        diff_from_overall_mean = mean_diff - overall_mean
        print(f"범주형 변수가 '{category}' 값을 가질 때, 그 그룹은 전체 평균과 {diff_from_overall_mean:.4f} 정도 차이가 납니다.")
```
